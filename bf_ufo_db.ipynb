```python
#Import Libraries

import pandas as pd
from sqlalchemy import create_engine
#extras
import os
```

```python
#Get UFO Data

csv_file = "Resources/ufo.complete.csv"

#Remove lines with errors

ufo_data_df = pd.read_csv(csv_file, error_bad_lines=False)
ufo_data_df.head(1)
```

```python
#Only keep columns we want

clean_ufo_data = ufo_data_df[['state', 'country', 'datetime','shape', 'duration (hours/min)', 'comments', 'latitude', 'longitude']].copy()
clean_ufo_data.head()
```

```python
# Estract info from CSV file.

csv_file = "Resources/clean_bf_reports.csv"
bf_data_df = pd.read_csv(csv_file)
bf_data_df.head(1)
```

```python
#clean up data to colums needed.

new_BF_data_df = bf_data_df[['state', 'title', 'date', 'latitude', 'longitude', 'location_details']].copy()
new_BF_data_df.head()
```

```python
#Merge the two data frames on Longitude

merge_df = new_BF_data_df.merge(clean_ufo_data, how='inner', on='longitude')
merge_df.head(4)
```

```python
#Rename the columns

merge_df_rename = merge_df.rename(columns={'state_x': 'Bigfoot State', 'title': 'Bigfoot Details', 'date': 'Bigfoot Date',
                                          'latitude_x': 'Bigfoot Latitude', 'location_details': 'UFO Details',
                                          'state_y': 'UFO State', 'country': 'UFO Country', 'datetime': 'UFO Date',
                                          'shape': 'UFO Shape', 'duration (hours/min)': 'UFO Sighting Duration', 'comments': 'UFO Comments',
                                          'latitude_y': 'UFO Latitude'})
merge_df_rename
```

```python
#Count number of rows to ensure all went smoothly

merge_df_rename.count()
```

```python
#Remove the extra commas to ensure that the csv will read into the database correctly.

cols = ['Bigfoot State', 'Bigfoot Details', 'Bigfoot Date', 'Bigfoot Latitude', 'longitude', 'UFO Details', 'UFO State',
       'UFO Country', 'UFO Date', 'UFO Shape', 'UFO Sighting Duration', 'UFO Comments', 'UFO Latitude']
merge_df_rename[cols] = merge_df_rename[cols].replace({',': ';'}, regex = True)
```

```python
merge_df_rename.head()
```

```python
#Create CSV

merge_df_rename.to_csv('Resources/bigfoot_ufo_longitude_csv_no_commas.csv')
```
